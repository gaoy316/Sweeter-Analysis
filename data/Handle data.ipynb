```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.read_csv('example1.csv')
```

```python
df2 = pd.read_csv('example2.csv')
```

```python
df3 = pd.read_csv('example3.csv')
```

```python
df4 = pd.read_csv('example4.csv')
```

```python
df5 = pd.read_csv('example5.csv')
```

```python
df=df1.append(df2).append(df3).append(df4).append(df5)
```

```python
df = df.drop('Unnamed: 0', 1)
```

```python
df.rename(columns = {'0':'Name'}, inplace = True)
df.rename(columns = {'1':'Class'}, inplace = True)
df.rename(columns = {'2':'SMILES'}, inplace = True)
df.rename(columns = {'3':'Class2'}, inplace = True)
df.rename(columns = {'4':'Calories'}, inplace = True)
df.rename(columns = {'5':'Sweetness'}, inplace = True)
df.rename(columns = {'6':'Origin'}, inplace = True)
df.rename(columns = {'7':'Approval'}, inplace = True)
df.rename(columns = {'8':'Therapeutic effect'}, inplace = True)
df.rename(columns = {'9':'Metabolism'}, inplace = True)
df.rename(columns = {'10':'Therapeutic annotation'}, inplace = True)
df.rename(columns = {'11':'Formula'}, inplace = True)
df.rename(columns = {'12':'Molweight'}, inplace = True)
df.rename(columns = {'13':'Atoms'}, inplace = True)
df.rename(columns = {'14':'Heavy Atoms'}, inplace = True)
df.rename(columns = {'15':'Rings'}, inplace = True)
df.rename(columns = {'16':'Bonds'}, inplace = True)
df.rename(columns = {'17':'Rotatable Bonds'}, inplace = True)
df.rename(columns = {'18':'H-bond Donors'}, inplace = True)
df.rename(columns = {'19':'H-bond Acceptors'}, inplace = True)
```

```python
df = df[pd.notnull(df['Name'])]
```

```python
df = df.reset_index(drop=True)
```
